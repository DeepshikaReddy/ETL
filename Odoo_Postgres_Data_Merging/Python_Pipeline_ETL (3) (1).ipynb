```python
import  pandas as pd
import psycopg2
import warnings
warnings.filterwarnings("ignore")

#Connection details of Odoo database
conn_odoo = psycopg2.connect(
    host="34.239.105.99",
    database="bitnami_odoo",
    user="postgres",
    password="Yd8pQqwdpzHL")
```

```python
#function to calculate start_time
def calculate_start_time():
    start_time = time.time()
    return start_time
    
```

```python
#function to calculate end_time
def calculate_end_time(start_time):
    end_time=time.time() - start_time
    return end_time
```

```python
import time

#fetch res_country data
cursor = conn_odoo.cursor()

start_time=calculate_start_time()

sql_country="SELECT name as Shipping_Country,id from res_country where name='US';"
df_country = pd.read_sql(sql_country,conn_odoo)

end_time_1=calculate_end_time(start_time)
response=f"<<Time taken to fetch res_country with  {len(df_country)} row>> "+str(end_time_1)+" seconds"
print('\033[1m'+response)
```

```python
#fetch res_country_state data

start_time=calculate_start_time()

country_id=df_country['id']
sql_state=f"SELECT id,name as Shipping_State,code,country_id from res_country_state where country_id={country_id.item()}"
df_state = pd.read_sql(sql_state,conn_odoo)

end_time_2=calculate_end_time(start_time)
response=f"<<Time taken to fetch res_country_state with  {len(df_state)} rows>> "+str(end_time_2)+" seconds"
print('\033[1m'+response)
```

```python
#function to convert list to comma separated string.
def list_to_str(list_val):
    list_str = [str(element) for element in list_val]
    list_str = ",".join(list_str)
    any_list_str="("+list_str+")"
    return any_list_str 
```

```python
#Creating State and Country dictionary
dict_state=pd.Series(df_state.shipping_state.values,index=df_state.id).to_dict()
dict_state
dict_country=pd.Series(df_country.shipping_country.values,index=df_country.id).to_dict()
dict_country
```

```python
#fetching res_partner data
start_time=calculate_start_time()

country_id=df_country['id']
state_list=df_state['id']
state_list_str=list_to_str(state_list)

sql_partner=f"SELECT id,name as Shipping_Name,email as Email,mobile as Phone,street as Shipping_Street,street2 as Shipping_Address1,city as Shipping_City,zip as Shipping_Zip,state_id,country_id from res_partner where country_id={country_id.item()} and state_id in {state_list_str}"
df_spartner = pd.read_sql(sql_partner,conn_odoo)

#fetch corresponding values of state and country by mapping from dictionary.
df_spartner['Shipping_Country']=df_spartner['country_id'].map(dict_country) 
df_spartner['Shipping_State']=df_spartner['state_id'].map(dict_state) 

#Sort by Partner-ID
df_spartner=df_spartner.sort_values(by=['id'])

end_time_3=calculate_end_time(start_time)
response=f"Result of tMap1<<Time taken to fetch res_partner with  {len(df_spartner)} rows >> "+str(end_time_3)+" seconds"
print('\033[1m'+response)
```
